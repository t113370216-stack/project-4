# Academic References: Recoding Rules & Data Cleaning Protocols
## Standardization for Multiple Logistic Regression Inference

### 🛠️ 1. Binary Standardization Rule / 二分變數標準化規範
To satisfy the rigorous assumptions of the Multiple Logistic Regression model, all core analytical variables are strictly dichotomized following the explicit project instruction protocol:
* **1 = Success / Yes / Exposed Group (具備該特徵或經歷高風險暴露)**
* **0 = Failure / No / Comparison Group (不具備該特徵或為對照基準組)**
* In the regression formula, `No Trauma` and `Male` are designated as the specific reference categories (Treatment contrast coding) to calculate corresponding Odds Ratios ($OR$) for trauma compounding effects and female vulnerability.

### ⚠️ 3. Missing Value Handling & Sample Constraints / 缺失值與樣本限制
* **Listwise Deletion**: Any observation containing missing, invalid, or skipped data codes across the target columns is completely purged via Pandas `.dropna()` prior to modeling. This guarantees the Logistic Regression runs on a perfectly aligned and cleanly encoded matrix.
* **Causal Escapement Boundary**: The analytical framework establishes cross-sectional associative odds at a static temporal horizon (2007). The tremendous Odds Ratios ($OR$) computed do not rigorously imply chronologically proven causal precedence, though they highlight a severe public health crisis.
