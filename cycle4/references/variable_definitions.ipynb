# Academic References: Variable Definitions & Benchmarks
## Data Source: CDC YRBS 2007 (Youth Risk Behavior Surveillance System)

### 📋 1. Response Variable / 依變量（反應變數）定義
* **Variable Name / 變數名稱**: `ConsideredSuicide` (Recoded to `Suicide_Binary` in analytical model)
* **Survey Question / 原始問句**: "During the past 12 months, did you ever seriously consider attempting suicide?" (過去12個月內，你是否曾認真考慮過嘗試自殺？)
* **Response Scale / 原始量尺**: Code 1 = Yes (Recoded to `1`), Code 2 = No (Recoded to `0`), Missing/Blank = Dropped.

### 🧬 2. Predictor Variables & Feature Engineering / 自變量與特徵工程定義

#### Predictor 1: Intimate Partner Violence / 親密關係暴力
* **Variable Name / 變數名稱**: `BoyfriendGirlfriendPhysicallyHurt` (Recoded to `PartnerViolence_Binary`)
* **Survey Question / 原始問句**: "During the past 12 months, did your boyfriend or girlfriend ever hit, slap, or physically hurt you on purpose?"
* **Response Scale / 原始量尺**: Code 1 = Yes (Recoded to `1`), Code 2 = No (Recoded to `0`).

#### Predictor 2: Forced Sexual Intercourse / 強迫性行為 (性創傷)
* **Variable Name / 變數名稱**: `ForcedSexualIntercourse` (Recoded to `ForcedSex_Binary`)
* **Survey Question / 原始問句**: "Have you ever been physically forced to have sexual intercourse when you did not want to?"
* **Response Scale / 原始量尺**: Code 1 = Yes (Recoded to `1`), Code 2 = No (Recoded to `0`).

#### Engineered Feature: Composite Trauma Level / 複合創傷指標
* **Variable Name / 變數名稱**: `Trauma_Level`
* **Engineering Logic / 建構邏輯**: Summation of `PartnerViolence_Binary` and `ForcedSex_Binary`.
  * **Both Traumas (雙重創傷)**: Score = 2. Exposed to both IPV and forced sex.
  * **Either Trauma (單一創傷)**: Score = 1. Exposed to exactly one form of trauma.
  * **No Trauma (無創傷)**: Score = 0. Baseline comparison group.

### ⚖️ 3. Control Variable / 控制變數定義

#### Control 1: Biological Sex / 生理性別
* **Variable Name / 變數名稱**: `WhatIsYourSex` (Mapped to `Sex_Label`)
* **Survey Question / 原始問句**: "What is your sex?"
* **Response Scale / 原始量尺**: Code 1 = Female, Code 2 = Male. Used as a baseline control to isolate the pure impact of trauma and explore the gender vulnerability gap in mental health.
