## PART I: METADATA OVERVIEW & MULTI-VARIABLE FRAMEWORK
### (一) 分析架構與基準設定

* **Response Variable (反應變數)**:

* **Primary Predictor (核心自變數)**: 
  * `Trauma_Level` (複合創傷指標):

    * **Either Trauma (單一創傷)**: 僅遭遇其中一種暴力。
    * **No Trauma (無創傷)**: 未遭遇上述暴力 *(Regression Baseline / 模型基準組)*。
* **Control Variable (控制變數)**: 
  * `Sex_Label` (生理性別): Female vs. Male *(Regression Baseline / 模型基準組)*。

## PART II: STATISTICAL INFERENCE & LOGISTIC REGRESSION RESULTS
### (二) 羅吉斯迴歸推論結果與勝算比 (Odds Ratios) 解析

藉由多元羅吉斯迴歸模型分離干擾變數後，所有核心預測因子的 $P$ 值皆達到極顯著水準 ($P < 0.001$)。

* **💥 The Compounding Effect of Trauma (創傷加乘效應)**:
  * **Both Traumas vs. No Trauma**: $OR \approx 6.00$
    * *詮釋*: 在控制性別基準後，遭受雙重創傷的青少年，其產生自殺意念的勝算（Odds）高達無創傷學生的 **6.0 倍**。這嚴格證明了創傷並非單純相加，而是產生了極具破壞性的加乘效應 (1+1 > 2)。
  * **Either Trauma vs. No Trauma**: $OR \approx 2.70$
    * *詮釋*: 即使僅遭遇單一創傷，青少年的自殺意念勝算依然達到基準組的 **2.7 倍**。

* **🧬 The Gender Gap in Mental Health (心理健康的性別落差)**:
  * **Female vs. Male Baseline**: $OR \approx 1.99$
    * *詮釋*: 數據揭示了嚴峻的性別不平等。在面臨完全相同等級的創傷（或同為無創傷）下，女性青少年考慮自殺的勝算是男性的 **1.99 倍（近乎兩倍）**，凸顯了女性在青春期的高度情感內化脆弱性。

  *鑑於雙重創傷導致自殺風險飆升 6 倍，校園輔導機制必須摒棄「單一事件單一處理」的思維。當通報親密關係暴力時，必須主動且敏感地篩檢潛在的性創傷史，以攔截毀滅性的心理崩潰。*

  The immutable baseline gender gap ($OR \approx 2.0$ for females) strongly argues against a "one-size-fits-all" intervention strategy. Public health frameworks must recognize the disproportionate social stigma and internalizing burden placed on young women, prioritizing gender-specific emotional regulation and empowerment programs.
  *女性在相同條件下高達兩倍的自殺勝算，強烈反對「一體適用」的干預策略。公衛體系必須正視年輕女性所承受的不成比例的社會污名與內化壓力，優先推動具備性別敏感度（Gender-sensitive）的情緒支持與賦權計畫。*
